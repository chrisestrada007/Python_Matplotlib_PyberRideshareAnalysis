```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)
# Combine the data into a single dataset
combined_data_df = pd.merge(city_data, ride_data, how="outer", on="city")
# Display the data table for preview
combined_data_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each city type
urban_df = combined_data_df.loc[combined_data_df["type"]=="Urban",:]
suburban_df = combined_data_df.loc[combined_data_df["type"]=="Suburban",:]
rural_df = combined_data_df.loc[combined_data_df["type"]=="Rural",:]
#-------------------------------------------------------------------
# Obtain the x and y coordinates for each city
city_groupby_df = combined_data_df.groupby("city")
total_rides_by_city = city_groupby_df["city"].count()
total_drivers_by_city= city_groupby_df["driver_count"].mean()
avg_city_fare = round(city_groupby_df["fare"].mean(),2)
#------------------------------------------------------------------
#urban values
urban_cities_df = urban_df.groupby("city")
total_rides_urban_city = urban_cities_df["city"].count()
total_drivers_urban_city= urban_cities_df["driver_count"].mean()
avg_urban_city_fare = round(urban_cities_df["fare"].mean(),2)
#------------------------------------------------------------------
#Suburban values
suburban_cities_df = suburban_df.groupby("city")
total_rides_suburban_city = suburban_cities_df["city"].count()
total_drivers_suburban_city= suburban_cities_df["driver_count"].mean()
avg_suburban_city_fare = round(suburban_cities_df["fare"].mean(),2)
#------------------------------------------------------------------
#Rural values
rural_cities_df = rural_df.groupby("city")
total_rides_rural_city = rural_cities_df["city"].count()
total_drivers_rural_city= rural_cities_df["driver_count"].mean()
avg_rural_city_fare = round(rural_cities_df["fare"].mean(),2)
#---------------------------------------------------------------------------
# Build the scatter plot
#all_city_scatter = plt.scatter(total_rides_by_city, avg_city_fare, marker="o", facecolors="red", 
#                           edgecolors="black", s=total_drivers_by_city, alpha=0.75)
urban_scatter = plt.scatter(total_rides_urban_city, avg_urban_city_fare, marker="o", facecolors="lightcoral", 
                edgecolors="black", s=total_drivers_urban_city*10, alpha=0.50, label="Urban")
suburban_scatter = plt.scatter(total_rides_suburban_city, avg_suburban_city_fare, marker="o", facecolors="lightskyblue", 
                edgecolors="black", s=total_drivers_suburban_city*10, alpha=0.50, label="Suburban")
rural_scatter = plt.scatter(total_rides_rural_city, avg_rural_city_fare, marker="o", facecolors="gold", 
                edgecolors="black", s=total_drivers_rural_city*10, alpha=0.50, label="Rural")
#---------------------------------------------------------------------------
# Incorporate the other graph properties
plt.suptitle("Pyber Ride Sharing Data (2016)", fontsize=16, fontweight="bold")
plt.ylabel("Average fare ($)")
plt.xlabel("Number of Rides")
#--------------------------------------------------------------------------
# Create a legend
plt.legend(title="City Types", fontsize='large')
#--------------------------------------------------------------------------
# Incorporate a text label regarding circle size
plt.text(41.5, 30, "Note:Circle size denotes amount of drivers per city.")
#--------------------------------------------------------------------------
# Save Figure
plt.savefig("images/pyber_data_2016.png")
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
total_urban_fares = round(urban_df["fare"].sum(),2)
total_suburban_fares = round(suburban_df["fare"].sum(),2)
total_rural_fares = round(rural_df["fare"].sum(),2)
total_fares = total_urban_fares +total_suburban_fares + total_rural_fares
# Calculate Type Percents
urban_fares_percent = total_urban_fares/total_fares
suburban_fares_percent = total_suburban_fares/total_fares
rural_fares_percent = total_rural_fares/total_fares
# Build Pie Chart
labels = combined_data_df["type"].unique()
sizes = [urban_fares_percent, suburban_fares_percent, rural_fares_percent]
colors = ["lightcoral","lightskyblue", "gold"]
explode = (.1, 0, 0)
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=90)
plt.suptitle("% Total Fares by City Type (2016)", fontsize=16, fontweight="bold")
# Save Figure
plt.savefig("images/pyber_fare_percents_2016.png")
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Total rides by type
total_urban_rides = len(urban_df["ride_id"])
total_suburban_rides = len(suburban_df["ride_id"])
total_rural_rides = len(rural_df["ride_id"])
total_rides = total_urban_rides +total_suburban_rides + total_rural_rides
#-------------------------------------------------------
# Calculate Ride Percents

#------------------------------------------------------------
# Calculate Type Percents
urban_rides_percent = total_urban_rides/total_rides
suburban_rides_percent = total_suburban_rides/total_rides
rural_rides_percent = total_rural_rides/total_rides
#-------------------------------------------------------------
# Build Pie Chart
labels = combined_data_df["type"].unique()
sizes = [urban_rides_percent, suburban_rides_percent, rural_rides_percent]
colors = ["lightcoral","lightskyblue", "gold"]
explode = (.1, 0, 0)
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=90)
plt.suptitle("% Total Rides by City Type (2016)", fontsize=16, fontweight="bold")
# Save Figure
plt.savefig("images/pyber_ride_percents_2016.png")
```

```python
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Total Drivers per City
total_drivers_urban_city= sum(urban_cities_df["driver_count"].mean())
total_drivers_suburban_city= sum(suburban_cities_df["driver_count"].mean())
total_drivers_rural_city= sum(rural_cities_df["driver_count"].mean())
total_drivers = total_drivers_urban_city + total_drivers_suburban_city + total_drivers_rural_city
# Calculate Driver Percents
percent_drivers_urban=total_drivers_urban_city/total_drivers
percent_drivers_suburban=total_drivers_suburban_city/total_drivers
percent_drivers_rural=total_drivers_rural_city/total_drivers
#-------------------------------------------------------------
# Build Pie Chart
labels = combined_data_df["type"].unique()
sizes = [percent_drivers_urban, percent_drivers_suburban, percent_drivers_rural]
colors = ["lightcoral","lightskyblue", "gold"]
explode = (.1, 0, 0)
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=90)
plt.suptitle("% Total Drivers by City Type (2016)", fontsize=16, fontweight="bold")
# Save Figure
plt.savefig("images/pyber_drivers_percents_2016.png")
# Three Insights:
print("Three Insights:")
print("1. Urban is by large the largest demographic in all regards.")
print("2. Suburban makes up 1/3 of sales for having just a little more than 1/6 of total drivers.")
print("3. Rural has the largest fares per ride, yet makeup a minimal amount of total sales.")
```
